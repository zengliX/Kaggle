```python
import h2o

# Start H2O on your local machine
h2o.init()

# Get help
#help(h2o.estimators.glm.H2OGeneralizedLinearEstimator)
#help(h2o.estimators.gbm.H2OGradientBoostingEstimator)
#help(h2o.estimators.deeplearning.H2ODeepLearningEstimator)
```

```python
from h2o.estimators.deeplearning import H2ODeepLearningEstimator
```

```python
train = h2o.import_file("./data/train.csv")
test = h2o.import_file("./data/test.csv")
```

```python
train.describe()
```

```python
cols_ignore = ['X11', 'X93', 'X107', 'X223', 'X235', 'X268', 'X289', 'X290', 'X293', 'X297', 'X330', 'X347']
```

```python
train = train.drop(cols_ignore)
```

```python
train.describe()
```

```python
# covert 0-1 columns to factor type
for i in range(10,366):
    train[i] = train[i].asfactor()
```

```python
x_train = train.names[2:]
y_train = train.names[1]
```

```python
model_cv = H2ODeepLearningEstimator(
    distribution = "gaussian",
    activation = "RectifierWithDropout",
    hidden = [256,128,64],
    input_dropout_ratio = 0.2,
    sparse = True,
    l1 = 1e-3,
    epochs = 10,
    nfolds = 5,
    variable_importances = True)
```

```python
model_cv.train(x = x_train, y = y_train, training_frame = train)
```

```python
model_cv.model_performance(train=True) 
```

```python
model_cv.varimp()
```

```python
pred = model_cv.predict(test)
```
